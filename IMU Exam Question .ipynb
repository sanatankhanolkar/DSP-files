```python
accX=Data["Accelerometer X (g)"].values     
```

```python
accY=Data["Accelerometer Y (g)"].values     
accZ=Data["Accelerometer Z (g)"].values     
```

```python
print(accX)
```

```python
gyroX=Data["Gyroscope X (deg/s)"].values
```

```python
gyroY=Data["Gyroscope X (deg/s)"].values
gyroZ=Data["Gyroscope X (deg/s)"].values
```

```python
import numpy as np
gyroX=np.deg2rad(gyroX)
gyroY=np.deg2rad(gyroY)
gyroZ=np.deg2rad(gyroZ)
```

```python
f=100
T=1/f
T
```

```python
g=9.81 # Conversion formula from gravity force to m/s2 1g =9.81 m/s2   1m/s2 =g/9.81
```

```python
AccX=accX/9.81
AccY=accY/9.81
AccZ=accZ/9.81
```

```python
# Calculate accelerometer  with new variable acc
acc=np.sqrt(AccX**2+AccY**2+AccZ**2)
```

```python
print(acc)
```

```python
len(acc)
```

```python
acc=np.array(acc)
```

```python
start_time = pd.to_datetime('10:00:00') # Generating Timestamps  using start time
timestamps = [start_time + pd.Timedelta(minutes=i) for i in range(len(acc))]
timestamps=np.array(timestamps)
```

```python
print(timestamps)
```

```python
# window size filter =10
# Moving average filter (mvf)
wsf=10
mvf=(np.ones(wsf))/(wsf)
```

```python
acc_smooth=np.convolve(mvf,acc)
```

```python
min_length=min(len(acc),len(acc_smooth))
```
